```python
def optimizing(inputs,model,obj):
    
    def optimize(X):   
        h=[X[Rotation],X[1],X[2][0],X[3][0],X[4],X[5],X[6],X[7],X[8]]

        objective1=[]
        for i in obj:
            if i!=6:
                target=model.predict(h)[i]
                objective1.append(-target)
            else:
                target=model.predict(h)[i]
                objective1.append(target)
                

        return objective1

    def x_var(Xdim,Ydim,Numberoffloor,RotationAngle,Wallreflectance,
              FloorReflectance,Ceilingreflectance,WWRnorth,WWReast,
              WWRsouth,WWRwest,GlassVisualtransmittance,shadertypenorth,
               shadertypeeast,shadertypesouth,shadertypewest,
               shaderdephtnorth, shaderdephteast, shaderdephtsouth,
               shaderdephtwest, shadercountnorth,shadercounteast,
               shadercountsouth,shadercountwest,shaderanglenorth,
               shaderangleeast, shaderanglesouth,shaderanglewest,
               SouthNeighbordistance,SouthNeighborheight,northNeighbordistance,
               northNeighborheight, WestNeighbordistance,WestNeighborheight,
               EastNeighbordistance,EastNeighborheight):

        variables0=Real(Xdim[0],Xdim[1])
        variables1=Real(Ydim[0],Ydim[1])

        variables2=Subset(Numberoffloor,1)
        variables3=Subset(RotationAngle,1)
        
        variables4=Real(Wallreflectance[0],Wallreflectance[1]) 
        variables5=Real(FloorReflectance[0],FloorReflectance[1])   
        variables6=Real(Ceilingreflectance[0],Ceilingreflectance[1])
        
        variables7=Real(WWRnorth[0],WWRnorth[1])   
        variables8=Real(WWReast[0],WWReast[1])
        variables9=Real(WWRsouth[0],WWRsouth[1])
        variables10=Real(WWRwest[0],WWRwest[1])
        
        variables11=Subset(GlassVisualtransmittance,1)
        
        variables12=Subset(shadertypenorth,1)
        variables13=Subset(shadertypeeast,1)
        variables14=Subset(shadertypesouth,1)
        variables15=Subset(shadertypewest,1)
        
        variables16=Real(shaderdephtnorth[0],shaderdephtnorth[1])
        variables17=Real(shaderdephteast[0],shaderdephteast[1])
        variables18=Real(shaderdephtsouth[0],shaderdephtsouth[1])
        variables19=Real(shaderdephtwest[0],shaderdephtwest[1])
        
        variables20=Subset(shadercountnorth,1)
        variables21=Subset(shadercounteast,1)
        variables22=Subset(shadercountsouth,1)
        variables23=Subset(shadercountwest,1) 
        
        
        variables24=Subset(shaderanglenorth,1)
        variables25=Subset(shaderangleeast,1)
        variables26=Subset(shaderanglesouth,1)
        variables27=Subset(shaderanglewest,1)
        
        variables28=Real(SouthNeighbordistance[0],SouthNeighbordistance[1])
        variables29=Real(SouthNeighborheight[0],SouthNeighborheight[1])
        variables30=Real(northNeighbordistance[0],northNeighbordistance[1]) 
        variables31=Real(northNeighborheight[0],northNeighborheight[1])
        variables32=Real(WestNeighbordistance[0],WestNeighbordistance[1])
        variables33=Real(WestNeighborheight[0],WestNeighborheight[1])
        variables34=Real(EastNeighbordistance[0],EastNeighbordistance[1])
        variables35=Real(EastNeighborheight[0],EastNeighborheight[1])

            

        return ([variables0,variables1,variables2,variables3,variables4,variables5,
                 variables6,variables7,variables8,variables9,variables10,
                 variables11,variables12,variables13,variables14,variables15,
                 variables16,variables17,variables18,variables19,variables20,
                 variables21,variables22,variables23,variables24,variables25,
                 variables26,variables27,variables28,variables29,variables30,
                 variables31,variables32,variables33,variables34,variables35])
    problem = Problem(36,len(obj))
    problem.types[:] =x_var(*inputs)
    problem.function = optimize

    algorithm = NSGAII(problem, variator=CompoundOperator(SSX(),SBX()))
    algorithm.run(1)

    p=[]
    p1=[]
    for solution in unique(nondominated(algorithm.result)):
        p.append(solution.variables)
        p1.append(solution.objectives)
        
    data1=pd.DataFrame(p,columns=['Xdim', 'Ydim', 'Numberoffloor', 'RotationAngle', 'Wallreflectance',
                                  'FloorReflectance', 'Ceilingreflectance', 'WWRnorth', 'WWReast',
                                  'WWRsouth', 'WWRwest', 'GlassVisualtransmittance', 'shadertypenorth',
                                   'shadertypeeast', 'shadertypesouth', 'shadertypewest',
                                   'shaderdephtnorth', 'shaderdephteast', 'shaderdephtsouth',
                                   'shaderdephtwest', 'shadercountnorth', 'shadercounteast',
                                   'shadercountsouth', 'shadercountwest', 'shaderanglenorth',
                                   'shaderangleeast', 'shaderanglesouth', 'shaderanglewest',
                                   'SouthNeighbordistance', 'SouthNeighborheight', 'northNeighbordistance',
                                   'northNeighborheight', 'WestNeighbordistance', 'WestNeighborheight',
                                   'EastNeighbordistance', 'EastNeighborheight'])
    
    data1=data1.explode(column=['Numberoffloor','RotationAngle','shadertypenorth','shadercountnorth','shaderanglenorth','shadertypeeast',
                                  'shadercounteast','shaderangleeast','shadertypesouth','shadercountsouth',
                                  'shaderanglesouth','shadertypewest','shadercountwest','shaderanglewest','GlassVisualtransmittance'])
    algorithm.result
    data1=round(data1,ndigits=2)
    data1=data1.iloc[:,:]
    s=list(model.predict(data1))
    data2=pd.DataFrame(s,columns=['mDA', 'UDI', 'UDI-Low', 'UDI-High', '1-sGA', 'sDA', 'ASE', 'cDA'])

    data3=pd.concat([data1,data2],axis=1)
    return data3
```
